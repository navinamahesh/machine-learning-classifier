# Initialization and Reading Data

```python
import csv
import json
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from collections import Counter
```

```python
# Path to data
CSV_PATH = "./wardrobe_items_downloaded_data.csv" # Path to wardrobe CSV data
EMILY_PATH = "./emily.json" # Path to dresses added to Emily's  list
TRISHA_PATH = "./trisha.json" #path to trisha.json file
```

```python
# reading in the csv data and both .json files
csv_data = pd.read_csv(CSV_PATH)

csv_data = csv_data.dropna()

emily_data = json.load(open(EMILY_PATH))
trisha_data = json.load(open(TRISHA_PATH))
```

# Data Visualization

## Visualizing Distribution of Dress Size

```python
#visualizing the data from the csv file:
#features = ['wardrobe_size','wardrobe_color','wardrobe_style','wardrobe_price','wardrobe_designer']
#wardrobe_id = 'wardrobe_id'

# Visualizing size:
size = dict(Counter(csv_data.wardrobe_size))
plt.bar(size.keys(),size.values())
plt.xlabel('size')
plt.show()
```

```python
#visualizing color:
color = dict(Counter(csv_data.wardrobe_color))
num_colors = len(color)
x_val = list(range(1,num_colors+1))
plt.bar(x_val,color.values(),align = 'center')
label = color.keys()
plt.xticks(x_val,label)
plt.xlabel('color')
plt.xticks(rotation = 90)
plt.show()
```

```python
#visualizing color:
style = dict(Counter(csv_data.wardrobe_style))
num_styles = len(style)
x_val = list(range(1,num_styles+1))
plt.bar(x_val,style.values(),align = 'center')
label = style.keys()
plt.xticks(x_val,label)
plt.xlabel('style')
plt.xticks(rotation = 90)
plt.show()
```
